```python
## get latitude and longitude for each team using the full_name

import requests
import pandas as pd
from tqdm import tqdm

# Load the CSV file into a DataFrame
df = ncaa_df

# Your Google API Key
API_KEY = 'AIzaSyCnVFNJ7I-Muja5_9tmWiP37nxLnmMrUQw'

# Define the function to get latitude and longitude
def get_lat_lon(place_name):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        'address': place_name,
        'key': API_KEY
    }
    response = requests.get(base_url, params=params)
    res_json = response.json()
    
    if res_json['status'] == 'OK':
        location = res_json['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        return None, None

df['Latitude'], df['Longitude'] = zip(*tqdm(df['full_name'].apply(get_lat_lon), total=df.shape[0]))


df.head()
```

```python
# save the dataframe
df.to_csv('../data/ncaa_com_fbs_fcs_table_with_location.csv', index=False)
```
